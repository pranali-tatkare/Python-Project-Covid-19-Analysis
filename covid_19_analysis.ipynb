```python
# 4. Check Latest Data by state
latest_data = data.sort_values('Date').groupby('State').tail(1)
print(latest_data[['State', 'Confirmed', 'Deaths', 'Recovered']])
```

```python
# 5. Top 10 Countries with Highest Cases
top_cases = latest_data.sort_values(by='Confirmed', ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x='Confirmed', y='State', data=top_cases, palette='Reds_r')
plt.title('Top 10 State with Highest COVID-19 Cases')
plt.xlabel('Total Cases')
plt.ylabel('State')
plt.show()
```

```python
data.columns
```

```python
# 6. Death Rate Calculation
latest_data['death_rate'] = (latest_data['Deaths'] / latest_data['Confirmed']) * 100
top_death_rate = latest_data.sort_values(by='death_rate', ascending=False).head(10)
print(top_death_rate[['State', 'death_rate']])
```

```python
# 7. Plot Death Rate Top 10
plt.figure(figsize=(12,6))
sns.barplot(x='death_rate', y='State', data=top_death_rate, palette='Blues_r')
plt.title('Top 10 State with Highest COVID-19 Death Rate (%)')
plt.xlabel('Death Rate (%)')
plt.ylabel('State')
plt.show()
```

```python
data.columns
```

```python
# 8. Recovered case Analysis
if 'Recovered' in data.columns:
    vaccinated = latest_data[['State', 'Recovered']].sort_values(by='Recovered', ascending=False).head(10)
    plt.figure(figsize=(12,6))
    sns.barplot(x='Recovered', y='State', data=vaccinated, palette='Greens_r')
    plt.title('Top 10 State by recovered case People')
    plt.xlabel('Recovered')
    plt.ylabel('State')
    plt.show()
```
